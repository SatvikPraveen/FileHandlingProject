- The project is based on `File Handling and String Manipulation`. It is to be performed on a text file which contains excerpts from a script of games of thrones. The text file is titled `conv.txt`
- There are two tasks with regards to the project: 
  - `First`: Find out the number of unique dialogue speakers in the sample conversation
  - `Second`: Create a new text file by the name of the dialogue speaker and store the unique words spoken by that character in the respective text file. Make sure there is only one word every line.

___
___

```python
# reading the conv.txt files
text_file = open('conv.txt', 'r')
lines = text_file.readlines()
print(lines)
```

```python
# finding total number of elements in the text_file
n = int(format(len(lines)))
print(n)
```

```python
# segregating speakers and their dialogues
dialogues = []
for i in range(0,n):
    if lines[i] != '\n':
        dialogues.append(lines[i].split(':'))
```

```python
dialogues
```

```python
# Total number of elements in dialogues.
k = int(format(len(dialogues)))
print(k)
```

```python
# Creating a list to put in all the speakers.
characters = []
for i in range(0,k):
    characters.append(dialogues[i][0])
```

```python
# Finding unique speakers.
unique_speakers = set(characters)
unique_speakers
```

# Task - First

Find the number of unique dialogue speakers in the sample conversation

```python
print("The number of unique dialogue speakers in the sample conversation is: " + str(len(unique_speakers)))
```

# Task - Second

Create a new text file by the name of the dialogue speaker and store the unique words spoken by that character in the respective text file. Make sure there is only one word every line.

```python
unique_characters = unique_speakers
```

```python
# Creating a list of unique characters(speakers) from the set.
unique_characters_list = []
for x in unique_characters:
    unique_characters_list.append(x)
```

```python
unique_characters_list
```

```python
# Creating a list containing all the dialogues of a particular speaker as elements.
d = []
character_wise_d = []
for x in unique_speakers:
    for i in range (0,k):
        if dialogues[i][0] == x:
            d.append(dialogues[i][1])
    print(d)
    character_wise_d.append(d)
    d = []
```

```python
# displaying the output of the list created above.
character_wise_d
```

```python
# Creating a unique word list using the list of the dialogues created above.
uniq_wd = []

for i in range(0, len(unique_speakers)):
    l = format(len(character_wise_d[i]))
    print(i,l)
    for j in range(0,int(l)):
        if j == 0:
            unique_words = set(character_wise_d[i][j].split(' '))
                           
        unique_words = unique_words.union(set(character_wise_d[i][j].split(' ')))
    unique_words = list(unique_words)
    s = format(len(unique_words))
    unique = []
    for m in range(int(s)):
        unique.append(unique_words[m]+ '\n')
    uniq_wd.append(unique)
```

```python
uniq_wd
```

```python
# creating dataframe.
import pandas as pd
df = pd.DataFrame(unique_characters_list)
df
```

```python
df.insert(1, "Dialogues", character_wise_d)
```

```python
df.insert(2, "Unique_Words", uniq_wd)
```

```python
column_names = ['Characters', 'Dialogues', 'Unique_Words']
```

```python
df.columns = column_names
```
